```python
plt.figure(figsize=(18,36))
for i in range(len(col)):
    plt.subplot(12,12,i+1)
    sns.barplot(stud[col[i]], color='green', orient='v')
    plt.tight_layout()
```

```python
plt.figure(figsize=(18,36))
for i in range(len(col)):
    plt.subplot(12,12,i+1)
    sns.boxplot(stud[col[i]], color='green', orient='v')
    plt.tight_layout()
```

```python
# there are outliers in maths and civics
```
